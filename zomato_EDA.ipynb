# EXPLORATORY DATA ANALYSIS AND VISUALIZATION ON ZOMATO DATA SET

# DATA CLEANING

```python
df.info()
```

```python
df.drop(['url', 'address','dish_liked','phone','menu_item','dish_liked',"reviews_list"],axis=1,inplace=True)
df
```

```python
df.head()
```

##### DROPPING DUPLICATES

```python
df.drop_duplicates(inplace=True)
df.shape
```

###### DROPPING NULL VALUES

```python
df.dropna(inplace=True)
df.head()
```

###### CLEANING RATE COLUMN

```python
df.rate.unique()
```

```python
def change(value):
    if(value=='-' or value=='NEW'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df["rate"]=df["rate"].apply(change);
df["rate"].head
```

```python
df["rate"].fillna(df["rate"].mean(),inplace=True)
```

```python
df.rate.isnull().sum()
```

```python
df.votes.unique()
```

```python
df.location.unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df.drop('listed_in(city)',axis=1,inplace=True)
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type'},inplace=True)
df.columns
```

###### CLEANING LOCATION COLUMN

```python
location=df.location.value_counts(ascending=False)
locationless300=location[location<300]
def change1(values):
    if(values in locationless300):
        return "others"
    else:
        return values
df["location"]=df["location"].apply(change1)
df.location.value_counts()
```

###### CLEANING REST_TYPE COLUMN

```python
type=df.rest_type.value_counts(ascending=True)
typeless600=type[type<600]
def update(value):
    if(value in typeless600):
        return "others"
    else:
        return value
df["rest_type"]=df["rest_type"].apply(update)
df.rest_type.value_counts()    
```

###### CLEANING CUISINES COLUMN

```python
var=df.cuisines.value_counts()
lessthan100=var[var<100]
def update1(values):
    if(values in lessthan100):
        return "others"
    else:
        return values
df["cuisines"]=df["cuisines"].apply(update1)
df.cuisines.value_counts()
```

###### CLEANING COST2PLATE COLUMN

```python
df.cost2plates.unique()
```

```python
def changes(values):
    if(',' in values):
        values=values.replace(",","")
        return float(values)
    else:
        return float(values)
df["cost2plates"]=df["cost2plates"].apply(changes)
df.cost2plates.unique()
```
